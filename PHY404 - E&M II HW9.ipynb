```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def reflection_coefficient(theta):
    
    a = np.sqrt(1 - ((n1/n2)*np.sin(theta*np.pi/180))**2)/(np.cos(theta*np.pi/180))
    
    b = n2/n1
    
    return ((a-b)/(a+b))**2


theta = np.linspace(0,100,1000)
n1 = 1
n2 = 1.5179
R = reflection_coefficient(theta)


def transmission_coefficient(theta):
    
    a = np.sqrt(1 - ((n1/n2)*np.sin(theta*np.pi/180))**2)/(np.cos(theta*np.pi/180))
    
    b = n2/n1
    
    C = a*b
    return C*(2/(a+b))**2


T = transmission_coefficient(theta)

plt.plot(theta,R,label = 'Reflection BK7.')
plt.plot(theta,T,label = 'Transmission BK7.')
plt.title('Reflection/Transmission Coefficients BK7 Glass')
plt.xlabel('theta(deg)')
plt.legend()
```
