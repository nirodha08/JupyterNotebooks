## Tungsten - Halogen Reference Spectrum

```python
ref = pd.read_csv('tun_ref.TXT', header=None, skiprows=0)

wav = ref[0][:-1].values
val = ref[1][:-1].values

wav = np.ndarray.tolist(wav)
val = np.ndarray.tolist(val)

wav = map(float,wav)
val = map(float,val)

wav = list(wav)
val = list(val)

#wav = [x**-9 for x in wav]

#plt.plot(wav, val)

def energy(lam):    ## Energy in Joules
    h = 6.62607015*10**-34 # plancks constant
    c = 299792458 #speed of light
    
    return [(h*c)/(lam*10**-9) for lam in lam]

E_ref = energy(wav)
E_ev = [E_ref * 6.241509*10**18 for E_ref in E_ref]

plt.plot(E_ev,val)
plt.xlabel('Photon Energy (eV)')
plt.ylabel('Reflectivity (a.u)')
plt.title('Tungsten - Halogen Reference Spectrum')
plt.show()
```
