```python
import numpy as np
import pandas as pd
from scipy import optimize
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import scipy
```

## Tungsten - Halogen Reference Spectrum

```python
ref = pd.read_csv('tun_ref.TXT', header=None, skiprows=0)

wav = ref[0][:-1].values
val = ref[1][:-1].values

wav = np.ndarray.tolist(wav)
val = np.ndarray.tolist(val)

wav = map(float,wav)
val = map(float,val)

wav = list(wav)
val = list(val)

#plt.plot(wav, val)

def energy(lam):
    h = 6.62607015**-34 # plancks constant
    c = 299792458 #speed of light
    
    return [(h*c)/(lam) for lam in lam]

E_ref = energy(wav)

plt.plot(E_ref,val)
```

### Sample 150 A - Room Temperature Reference

```python
rm_150 = pd.read_csv('150_room_temp.TXT', header=None, skiprows=0)

w_150 = rm_150[0][:-1].values
v_150 = rm_150[1][:-1].values

w_150 = np.ndarray.tolist(w_150)
v_150 = np.ndarray.tolist(v_150)

w_150 = map(float,w_150)
v_150 = map(float,v_150)

w_150 = list(w_150)
v_150 = list(v_150)

plt.plot(E_ref[1:], v_150[1:])
```
